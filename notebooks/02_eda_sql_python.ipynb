```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("../database/database.db")
```

```python
#load tables
sales = pd.read_sql_query("SELECT * FROM Sales", conn)
external = pd.read_sql_query("SELECT * FROM External_Factors", conn)
```

```python
#total sales
pd.read_sql_query("""
SELECT SUM(weekly_sales) AS total_sales
FROM Sales;
""", conn)
```

```python
#top stores
pd.read_sql_query("""
SELECT store_id, SUM(weekly_sales) AS total
FROM Sales
GROUP BY store_id
ORDER BY total DESC;
""", conn)
```

```python
#monthely trend
pd.read_sql_query("""
SELECT SUBSTR(date,1,7) AS month,
       SUM(weekly_sales) AS sales
FROM Sales
GROUP BY month
ORDER BY month;
""", conn)
```

```python
#economic impact
pd.read_sql_query("""
SELECT 
    s.store_id,
    AVG(s.weekly_sales) AS avg_sales,
    AVG(e.unemployment) AS avg_unemployment
FROM Sales s
JOIN External_Factors e
ON s.store_id = e.store_id AND s.date = e.date
GROUP BY s.store_id;
""", conn)
```

### Insight:
Stores with higher unemployment rates tend to show lower average sales, indicating a negative correlation between economic conditions and retail performance.
